# DA-CODING CHALLENGE

```python
#A little data cleaning
def clean_data(df):
  # remain with the 2 needed columns
  filtered_data_df = df[["communities_villages", "water_functioning"]]

  # replace "na_dn" values with "no", we can't drop the column since its a valid waterpoint
  filtered_data_df['water_functioning'].replace({'na_dn': 'no'}, inplace=True)

  return filtered_data_df
```

```python
#Sort out broken percentages
def get_broken_percentage(filtered_data_df, no_of_waterpoints):
  # The rank for each community by the percentage of broken water points.
  broken_df = filtered_data_df[filtered_data_df['water_functioning'] == 'no']
  broken_waterpoints = broken_df["communities_villages"].value_counts()

  # populate percentages
  results = {}

  for location in dict(no_of_waterpoints):
    if location in broken_waterpoints:
      results[location] = (broken_waterpoints[location] / no_of_waterpoints[location]) * 100
    else:
      results[location] = 0

 # sort ranking  percentages in descending order
results = sorted(results.items(), key=lambda x: x[1], reverse=True)
  return results
```

# Manipulation and solution

```python
#Import the data analysis and manipulation tool

import pandas as pd
#read data from url
url = 'https://raw.githubusercontent.com/onaio/ona-tech/master/data/water_points.json'

def calculate(url):
  data_df = pd.read_json(url)

  filtered_data_df = clean_data(data_df)
  
  # The number of water points that are functional can be found by counting the functional ones
  functional_waterpoints = (filtered_data_df['water_functioning'] == 'yes').sum()

  # The number of water points per community,
  no_of_waterpoints = filtered_data_df["communities_villages"].value_counts()

  # get percentage of broken
  broken_waterpoints_percentage = get_broken_percentage(filtered_data_df, no_of_waterpoints)
  
  result = {}

  result["number_functional"] = functional_waterpoints
  result["number_water_points"] = dict(no_of_waterpoints)
  result["community_ranking"] = broken_waterpoints_percentage

  return result
 
# driver code
calculate(url)
```
